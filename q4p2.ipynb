```python
import numpy as np

for i in range(3):
    print(f"\nMatrix {i+1}:")

    A = np.random.randint(-10, 11, (3,3))
    print(A)

    det = np.linalg.det(A)
    print("Determinant:", det)

    if det == 0:
        print("This matrix does not have an inverse because its determinant is 0.")
    else:
        A_inv = np.linalg.inv(A)
        print("Inverse:")
        print(A_inv)

        product = A @ A_inv
        print("A * A_inv:")
        print(np.round(product, 6))
```
